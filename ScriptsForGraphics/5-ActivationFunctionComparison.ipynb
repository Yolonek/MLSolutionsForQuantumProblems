```python
import numpy as np
from matplotlib import pyplot as plt
from functools import partial
```

```python
def step(x):
    return np.where(x >= 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def snake(x, alpha=1):
    return x + np.sin(alpha*x)**2 / alpha
```

```python
def adjust_ticks_and_grid(ax):
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels([])
    ax.tick_params(axis='x', length=0)
    ax.grid(True, which='both', axis='x')
    ax.set_xlim(x[0], x[-1])  
```

```python
a = 1.
figure, axes = plt.subplots(4, 1, layout='constrained', figsize=(3, 6))
functions = [step, sigmoid, relu, partial(snake, alpha=a)]
func_names = ['a) step', 'b) sigmoid', 'c) relu', f'd) snake($\\alpha={a}$)']
x = np.linspace(-5, 5, 400)
for function, function_name, ax in zip(functions, func_names, figure.axes):
    ax.plot(x, function(x), color='black', linewidth=2)
    ax.set(title=function_name)
    ax.grid()
    if ax != figure.axes[-1]:
        adjust_ticks_and_grid(ax)
figure.savefig(f'./ThesisGraphics/ActivationFunctions.png')
```
